Sales And Customer DataSet Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#uploaded the csv file into the canva

df = pd.read_csv("powerbi_practice_dataset.csv")
df.head()
```

```python
df.describe()  #basic statistics according to data
```

```python
#sales anylasis according to productcategory
df.groupby("ProductCategory")["SalesAmount"].sum().sort_values(ascending=False)  #This tells which category sells the most (Electronics, Grocery, Furniture, etc.)
```

```python
#showing with graph
df.groupby("ProductCategory")["SalesAmount"].sum().sort_values(ascending=False).plot(kind="bar",color="skyblue")
plt.title("Sales Anylasis (Sales vs Product)")
plt.xlabel("Product")
plt.ylabel("Sales Amount")
plt.legend()
plt.show()
```

```python
#sales anylasis according to subcategory
df.groupby("SubCategory")["SalesAmount"].sum().sort_values(ascending=False)  #This tell which subcategory sell the most
```

```python
#profit anylasis Depend on the Region
df.groupby("Region")["Profit"].sum().sort_values(ascending=False)  #Company should focus more on the top profit region.
```

```python
#profit anylasis Depend on the Region
df.groupby("Region")["Profit"].sum().sort_values(ascending=False).plot(kind="pie",color=["red","green","blue","yellow"],autopct="%1.1f%%")
plt.title("Profit Anylasis")
plt.ylabel("Profit")
plt.show()
```

```python
#coustomer anylasis
df["CustomerType"].value_counts()  #Shows customer loyalty and retention.
```

```python
#For customer count analysis, countplot is the most suitable seaborn chart because it directly shows the frequency of each customer category
sns.countplot(x="CustomerType",data=df,palette="dark",hue="Gender")
plt.xlabel("Customer Type")
plt.ylabel("Number of Customer")
plt.title("Customer Anylasis")
plt.show()
```

```python
#Gender-Based Profit anylasis
df.groupby("Gender")["Profit"].sum()
```

```python
# Why didnt you use groupby for violin plot?
# Violin plots show data distribution, so I used raw profit values instead of aggregated sums
sns.violinplot(x="Gender",y="Profit",data=df,inner="box",palette="rocket",hue="Gender")
plt.title("Profit Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Profit")
plt.show()
```

```python
#Gender-Based sales anylasis
df.groupby("Gender")["SalesAmount"].sum()
```

```python
#This is a categorical box plot used to compare the distribution of SalesAmount across Gender categories and to identify median values and outliers.
sns.boxplot(x="Gender",y="SalesAmount",data=df,palette="Set2",hue="Gender")
plt.title("SalesAmount Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Sales Amount")
plt.show()
```

```python
#Discount impact
df.groupby("DiscountPercent")["Profit"].mean()  #You can see how profit changes with discount.
```

```python
sns.barplot(x="DiscountPercent",y="Profit",data=df,palette="dark",hue="Gender")
plt.title("Average Discount Percent By Profit")
plt.show()
```

```python
#Delivery Performance
df.groupby("SalesPerson")["DeliveryDays"].mean().sort_values(ascending=False)
```

```python
#Return Status
df.groupby("ReturnStatus")["ProductCategory"].value_counts()
```

```python
sns.catplot(data=df,x="ReturnStatus",y="ProductCategory",kind="violin",hue="Gender",palette="Set2")
```

```python
df["CustomerRating"].value_counts()
```

```python
sns.displot(df,x="CustomerRating",palette="flare",hue="Gender",col="Gender") #discrete=True  xaxis ka point ko int me convert kar deata h
plt.title("Customer Rating Analysis")
plt.xlabel("Customer Rating")
plt.ylabel("Number of Counts")
plt.show()
```

```python
df.groupby("ProductCategory")["CustomerRating"].mean().plot(kind="line")
plt.title("Customer wise Product")
plt.xlabel("Product")
plt.ylabel("Rating")
plt.legend()
plt.grid(color="green")
plt.show()
```
